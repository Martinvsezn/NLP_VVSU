# Задание 3. MI и t-score
Выделение однословных и неоднословных терминов из загруженных текстов с использованием одной из мер терминологичности (MI, t-score, weirdness).

MI - способ проверить степень независимости появления двух слов в
тексте

- *Если слова полностью независимы, то вероятность их
совместного появления равна произведению вероятностей появления каждого из них (частот), а значение меры MI равно нулю.*

T-score - учитывает частоту совместной встречаемости ключевого слова и его коллоката

- *Отвечает на вопрос: насколько неслучайной является сила ассоциации (связанности) между коллокатами?*

Weirdness -  отношение частот в тематической и контрастивной коллекции

```python
# Извлекаем коллокации
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
import pymorphy3
import re
import numpy as np
```

```python
morph = pymorphy3.MorphAnalyzer()
```

```python
# Функция нормализации
def normalizePymorphy(text):
    tokens = re.findall('[а-яёА-ЯЁ]+-[а-яёА-ЯЁ]+-[а-яёА-ЯЁ]+|[а-яёА-ЯЁ]+-[а-яёА-ЯЁ]+|[а-яёА-ЯЁ]+', text)
    words = []
    for t in tokens:
        pv = morph.parse(t)
        if str(pv[0].tag.POS) in ['ADJF', 'ADJS', 'NOUN', 'VERB', 'PRTF', 'PRTS', 'GRND', 'PREP']:
            words.append(pv[0].normal_form)
    return words
```

```python
cut_articles = articles[:1000]
```

```python
# Нормализация статей
articles_norm = [' '.join(normalizePymorphy(a)) for a in tqdm(articles)]
```

```python
# Инициализация векторайзеров
tfv1 = CountVectorizer(token_pattern="[а-яёА-ЯЁ]+-[а-яёА-ЯЁ]+-[а-яёА-ЯЁ]+|[а-яёА-ЯЁ]+-[а-яёА-ЯЁ]+|[а-яёА-ЯЁ]+",
                      ngram_range=(1, 1)) #только униграммы
tfv2 = CountVectorizer(token_pattern="[а-яёА-ЯЁ]+-[а-яёА-ЯЁ]+-[а-яёА-ЯЁ]+|[а-яёА-ЯЁ]+-[а-яёА-ЯЁ]+|[а-яёА-ЯЁ]+",
                      ngram_range=(2, 2)) #только биграммы
```

```python
# Трансформация документов в числовые векторы
tf1_articles = tfv1.fit_transform(tqdm(articles_norm))
tf2_articles = tfv2.fit_transform(tqdm(articles_norm))
```

```python
list(tfv1.vocabulary_.items())[:50]
```

```python
list(tfv2.vocabulary_.items())[:50]
```

```python
#функция для рассчёта MI и t-score
def calc_MI_tscore(tf1_articles, tf2_articles, tfv1, tfv2, thr):
    mis = {}
    t_scores = {}
    corp_len = tf1_articles.sum()
    for pair, index2 in tqdm(tfv2.vocabulary_.items()):
        freq12 = tf2_articles[0, index2]
        if freq12 < thr:
            continue
        words = pair.split(' ')
        freq1 = tf1_articles[0, tfv1.vocabulary_[words[0]]]
        freq2 = tf1_articles[0, tfv1.vocabulary_[words[1]]]
        mis[pair] = freq12 / (freq1 * freq2)
        t_scores[pair] = (freq12 - (freq1 * freq2) / corp_len) / np.sqrt(freq12)
    return mis, t_scores
```

```python
# Рассчет MI и t-score
mis, t_scores = calc_MI_tscore(tf1_articles, tf2_articles, tfv1, tfv2, 4)
```

```python
sorted(mis.items(), key=lambda x: x[1], reverse=True)[:100]
```

```python
sorted(t_scores.items(), key=lambda x: x[1], reverse=True)[:100]
```

```python
# Запись в csv файлы
with open('mi_tech.csv', mode='w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerow(['Token', 'Weirdness'])
    for Token, MI in sorted(mis.items(), key=lambda x: x[1], reverse=True):
        writer.writerow([Token, MI])

with open('tscore_tech.csv', mode='w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerow(['Token', 'Weirdness'])
    for Token, TScore in sorted(t_scores.items(), key=lambda x: x[1], reverse=True):
        writer.writerow([Token, TScore])
```

# Задание 3. Weirdness

Выделение однословных и неоднословных терминов из загруженных текстов с использованием weirdness

```python
# Расчёт weirdness
weir_1 = {}
weir_2 = {}
for word, index in tqdm(tfv1.vocabulary_.items()):
    if word in med_tfv1.vocabulary_.keys():
        w = tf1_articles[0, index] / (med_tf1_articles[0, med_tfv1.vocabulary_[word]] + 1)
    else:
        w = tf1_articles[0, index]
    weir_1[word] = w

for word, index in tqdm(tfv2.vocabulary_.items()):
    if word in med_tfv2.vocabulary_.keys():
        w = tf2_articles[0, index] / (med_tf2_articles[0, med_tfv2.vocabulary_[word]] + 1)
    else:
        w = tf2_articles[0, index]
    weir_2[word] = w
```
